```python
# Replace '?' with NaN
df.replace('?', np.nan, inplace=True)
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Convert 'normalized-losses' to numeric
df['normalized-losses'] = pd.to_numeric(df['normalized-losses'])

# Impute missing values in 'normalized-losses' with the mean of the column
mean_normalized_losses = df['normalized-losses'].mean()
df['normalized-losses'].fillna(mean_normalized_losses, inplace=True)

print("Imputed missing values in 'normalized-losses' with the mean.")
```

```python
df.info()
```

```python
# Convert 'price' column to numeric
df['price'] = pd.to_numeric(df['price'])

# Drop rows where 'price' is NaN
df = df.dropna(subset=['price'])

print("Dropped rows with missing values in 'price' column.")
```

```python
# Boxplot for 'normalized-losses'
sns.boxplot(x=df['normalized-losses'])
```

```python
# Boxplot for 'symboling'
sns.boxplot(x=df['symboling'])
```

```python
# Define a function to remove outliers based on IQR
def remove_outliers(column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Remove outliers from 'normalized-losses' and 'symboling'
df = remove_outliers('normalized-losses')
df = remove_outliers('symboling')

print("Outliers removed from 'normalized-losses' and 'symboling'.")
```
